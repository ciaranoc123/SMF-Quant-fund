```python
import rank2 as rank2
import pandas as pd
import smf2021.Optimiser as O
```

```python
# Fields used for different industries / subsectors

# Industrials
ind_fields = ['TR.PE',"TR.PriceToBVPerShare", "TR.EPSMean", "TR.EVToSales", "TR.EVToEBIT",
              "TR.PtoEBTMeanEst", "TR.FwdPEG"]


# Comms
comms_fields = ['TR.ROATotalAssetsPercent', 'TR.ReturnonAVGCommEqtyPctIncomeAvailabletoCommExclExtraItems', 'TR.EVToEBITDA',
                'TR.ROICActValue', 'TR.TimesInterestEarned', 'TR.FCFMeanYield', ]

# Healthcare
health_fields = ['TR.PE', 'TR.EVToSales', 'TR.EVToEBITDA', 'TR.EVToSales', 'TR.ROICActValue', 'TR.FCFMeanYield',
'TR.ReturnonAVGCommEqtyPctIncomeAvailabletoCommExclExtraItems',"TR.PriceToBVPerShare"]

cons_fields = ['TR.PE', 'TR.ROATotalAssetsPercent', 'TR.ReturnonAVGCommEqtyPctIncomeAvailabletoCommExclExtraItems',
               'TR.EVToEBITDA', 'TR.EVToSales']
# Tech
#tech_fields = ['TR.PE', "TR.PriceToBVPerShare", 'TR.ROICActValue', 'TR.ReinvestmentRatePct','TR.ReturnonAVGCommEqtyPctIncomeAvailabletoCommExclExtraItems', 'TR.FCFMeanYield']
```

```python
# Ranker for industrials
ind_ranker = rank2.ranker(ind_fields)


# Rankers for comms
comms_ranker = rank2.ranker(comms_fields)

health_ranker = rank2.ranker(health_fields)

cons_ranker = rank2.ranker(cons_fields)

# Rankers for tech
#tech_ranker = rank2.ranker(tech_fields)
```

```python
all_comms = comms_ranker.get_data()
#all_tech = tech_ranker.get_data()
all_ind = ind_ranker.get_data()

all_health = health_ranker.get_data()

all_cons = cons_ranker.get_data()
```
